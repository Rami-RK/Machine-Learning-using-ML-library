```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
X2D=pca.fit_transform(X)
X2D
```

```python
##Explained Variance Ratio
pca.explained_variance_ratio_
```

The first dimension explains 84.2% of the variance, while the second explains 14.6%.
By projecting down to 2D, we lost about 1.1% of the variance:

```python
## plotting  our 3D data :>   X
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(X[:,0], X[:,1],X[:,2])
```

```python

## Plotting the plane on which data is being projected to make 2d
fig = plt.figure()
ax = plt.axes(projection='3d')
axes = [-1.8, 1.8, -1.3, 1.3, -1.0, 1.0]
x1s = np.linspace(axes[0], axes[1], 10)
x2s = np.linspace(axes[2], axes[3], 10)
x1, x2 = np.meshgrid(x1s, x2s)
C = pca.components_ ## It gives direction of plane
R = C.T.dot(C)
z = (R[0, 2] * x1 + R[1, 2] * x2) / (1 - R[2, 2]) # Equation of plane when normal to the plane is known
ax.plot_surface(x1, x2, z, alpha=0.2, color="k")
```

#### Implementing PCA   for compression of  image data of digit mnist

```python
mnist=np.load('mnist_data.npy').item()
```

```python
mnist.keys()
```

```python
X, y = mnist["data"], mnist["target"]
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
sample_digit = X_train[28]
sample_digit_image = sample_digit.reshape(28, 28)
plt.imshow(sample_digit_image, cmap=mpl.cm.binary)
plt.show()
```

```python
pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
d
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
xx=np.arange(1,785)
xx
```

```python
plt.plot(xx,cumsum)
```

```python
len(cumsum)
```

```python
pca=PCA(n_components=154)
X_reduced=pca.fit_transform(X_train)
```

 Now apply all your classification operation  in 'X_reduced' data. While most of the variance(95%) is preserved, the dataset is now less than 20% of its original size.This speed up classification algorithm tremendously.

##### Transform data back to its original space.

```python
X_recovered = pca.inverse_transform(X_reduced)
```

```python
sample_digit = X_recovered[28]
sample_digit_image = sample_digit.reshape(28, 28)
plt.imshow(sample_digit_image, cmap=mpl.cm.binary)
plt.show()
```
