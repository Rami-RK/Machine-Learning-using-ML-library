```python
#Ramendra Kumar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('ex1data2.txt',header=None)
X=data.iloc[:,[0,1]]
Y=data.iloc[:,[2]]            
```

```python
data.head()# first column size of flat,second column no of bedrooms,third column price of the flat
```

```python
X
```

```python
Y
```

```python
X=np.array(X)

#X=(X-np.mean(X))/np.std(X)#Normalizatin & Feature scaling
Y=np.array(Y)
m=len(Y)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X,Y,test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, Y_train)
```

```python
reg.coef_ 
```

```python
reg.intercept_ 
```

```python
y_pred=reg.predict(X_test)
```

```python
y_pred
```

```python
reg.score(X_test,Y_test)
```

```python
###
m=len(Y_test)
X_axis=np.arange(1,m+1)
y_pred=reg.predict(X_test)
plt.plot(X_axis,Y_test,'k-') # orginal
plt.plot(X_axis,y_pred,'r-')
plt.show()
```
